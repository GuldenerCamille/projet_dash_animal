```python
import plotly.graph_objects as go
import plotly.express as px
import pandas as pd

df = pd.read_csv("steam.csv")
df['release_date'] = pd.to_datetime(df['release_date'])


df.columns
```

```python
df
```

```python
date_chose = 2017
# var_sort = "owners"
df_date = df[df['release_date'].dt.year== date_chose]
df_date = df_date.sort_values(['owners', 'median_playtime'],
              ascending = [False, False])
fig_date_chose = go.Figure(data=[go.Table(
    header=dict(values=["Name","Editeur","Prix","Genres","Poce bleu"],
                fill_color='paleturquoise',
                align='left'),
    cells=dict(values=[df_date.name,df_date.developer,
                       df_date.price,df_date.genres,df_date.positive_ratings],
               fill_color='lavender',
               align='left'))
])
fig_date_chose.show()
```

```python
fig =px.histogram(df, x=df["release_date"].dt.year)
fig.update_layout(
    title_text='Game per year', # title of plot
    xaxis_title_text='Year', # xaxis label
    yaxis_title_text='Count', # yaxis label
    bargap=0.2, # gap between bars of adjacent location coordinates
    bargroupgap=0.1 # gap between bars of the same location coordinates
)
```
